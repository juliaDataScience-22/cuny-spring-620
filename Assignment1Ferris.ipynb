Below, a graph is created to show the nodes of the Krackhardt kite with edges. The names are in order alphabetically from one side to the other. To create the graph, the packages networkx and matplotlib were used.

First, the graph was created and set as the variable G. Next, the labels are set as the names. To create the graph, the function spring_layout() was used next. Other functions were used to make the nodes. Features like square nodes and black writing were added using parameters in those functions. The labels were added below the nodes with a for loop. Then, to add a border around the whole graph, another for loop was used for each side of the graph. Some functions from matplotlib were used to display the graph to the screen.



```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.krackhardt_kite_graph()

labels = {
    0: 'Beverly', 1: 'Andre', 2: 'Ed', 3: 'Diane', 4: 'Carol',
    5: 'Garth', 6: 'Fernando', 7: 'Heather', 8: 'Ike', 9: 'Jane'
}

pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=2000, node_shape='s', edgecolors='black', linewidths=2)
nx.draw_networkx_edges(G, pos, edge_color='gray')

for node, (x, y) in pos.items():
    plt.text(x, y - 0.1, labels[node], fontsize=12, ha='center', color='black')

plt.title("Krackhardt Kite Graph")
plt.axis('off')
plt.show()
```
